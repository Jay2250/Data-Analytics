```python
# Scatter plot of age versus resting blood pressure
df = pd.read_csv("heart.csv")

print(df.head())
print(df.info())
```

```python
# Detailing using color coding - Matplotlib version
# marker="^": Specifies the marker style for the scatter plot.
# s=60: Sets the size of the markers to 60 points.
# c=df["age"]: Specifies the color of the markers based on the values in the 'age' column of the DataFrame df. The color of each marker corresponds to the respective age value.
# edgecolors="k": Sets the color of the marker edges to black ('k').
# alpha=0.5: Sets the transparency of the markers to 0.5 (semi-transparent).
fig, ax = plt.subplots()
ax.scatter(df["age"], df["trestbps"], marker="^", 
           s=60, c=df["age"], edgecolors="k", alpha=0.5)
ax.set_xlabel("age")
ax.set_ylabel("trestbps");
plt.show()
```

```python
x = np.arange(0,10)
y = 2 * x

print(x)
print(y)
```

```python
# Creating a Figure

# The main idea in using the more formal Object Oriented method is to create a
#   figure object and then just call methods or attributes off of that object -
#   This approach is nicer when dealing with a canvas that has multiple plots on it
```

```python
# Create an empty canvas
fig = plt.figure()
# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)
# Plot on that set of axes
axes.plot(x, y)
plt.show()
```

```python
# Create Figure (empty canvas)
fig = plt.figure()
# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)
# Plot on that set of axes
axes.plot(a, b)
plt.show()
```

```python
# Adding another set of axes to the Figure
# So far we have only seen one set of axes on this figure object, 
#   but we can keep adding new axes on to it at any location and size we want
#   We can then plot on that new set of axes
# Code is a little more complicated, but the advantage is that we now 
#   have full control of where the plot axes are placed, and we can easily 
#   add more than one axis to the figure
#   Note how we are plotting a, b twice here
# Create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Large figure
axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5]) # Smaller figure

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_title('Small Figure');

plt.show()
```

```python
# Let's move the small figure and edit its parameters

# Create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Large figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Smaller figure

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');

plt.show()
```

```python
# KDE (Kernel Density Estimation) plot
# Maps an estimate of a probability *density* function of a random variable
#  Kernel density estimation is a fundamental data smoothing problem where 
#  inferences about the population are made, based on a finite data sample
np.random.seed(42)
```

```python
# randint should be uniform, each age has the same chance of being chosen
#   note: in reality ages are almost never uniformally distributed, but this 
#   is just an example
sample_ages = np.random.randint(0,100,200)
print(sample_ages)
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=["age"])
print(sample_ages.head())

sns.rugplot(data=sample_ages,x='age')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',bins=10,rug=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',bins=10,rug=True,kde=True)
plt.show()
```

```python
sns.kdeplot(data=sample_ages,x='age')
plt.show()
```

```python
# Cut-off KDE 
#  We could cut off the KDE if we know our data has hard limits 
#   (no one can be a negative age and no one in the population can be older than 100 
#    for some reason)
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(data=sample_ages,x='age',clip=[0,100])
plt.show()
```

```python
# Basic styling of KDE plot
sns.kdeplot(data=sample_ages,x='age',bw_adjust=0.5,fill=True,color='red')
plt.show()
```

```python
df = pd.read_csv("StudentsPerformance.csv")
print(df.head())
```

```python
# Box plot
# Display distribution through the use of quartiles and an IQR for outliers
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df)
plt.show()
```

```python
# Add hue for further segmentation
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df,hue='gender')
plt.show()
```

```python
# changing orientation
sns.boxplot(x='math score',y='parental level of education',data=df,orient='h')
plt.show()
```

```python
# Violin plot
# Plays a similar role as a box and whisker plot - It shows the distribution 
#   of quantitative data across several levels of one (or more) categorical variables 
#   such that those distributions can be compared
#   Unlike a box plot, in which all of the plot components correspond 
#   to actual datapoints, the violin plot features a kernel density estimation 
#   of the underlying distribution
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,hue='gender')
plt.show()
```

```python
# split
# When using hue nesting with a variable that takes two levels, 
#   setting split to True will draw half of a violin for each level
#   This can make it easier to directly compare the distributions
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,hue='gender',split=True)
plt.show()
```

```python
# inner
# Representation of the datapoints in the violin interior
#   If box, draw a miniature boxplot
#   If quartiles, draw the quartiles of the distribution
#   If point or stick, show each underlying datapoint
#   Using None will draw unadorned violins
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,inner=None)
plt.show() 
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,inner='box')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,inner='quartile')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,inner='stick')
plt.show()
```

```python
# orientation
# Simply switch the continuous variable to y and the categorical to x
sns.violinplot(x='math score',y='parental level of education',data=df,)
plt.show()
```

```python
# bandwidth
# Similar to bandwidth argument for kdeplot (controls the smoothness)
#   Low value = Low smoothness, High value = High smoothness

plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df,bw=0.1)
plt.show()
```
